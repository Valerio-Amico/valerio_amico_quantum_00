```python
nome="fid_vs_steps_simu_iterated_nm.txt"

f=open("dati_esame_ruben/"+nome, "r")

steps=[]
dati=[]
dati_dev=[]

for x in f:
    if x[0]!="#":
        y=x.split(" ")
        steps.append(float(y[0]))
        dati.append(float(y[1]))
        dati_dev.append(float(y[2]))
```

```python
plt.errorbar(steps, dati,dati_dev,linestyle=':',marker="*",label="vecchia calibrazione")

plt.plot([0,100],[1,1],color="gray")
plt.xlabel('steps')
plt.ylabel('fidelity')
plt.title('fidelity vs steps')
plt.legend()
plt.grid()

plt.show()
```
